```python
# 서울시 인구 현황 데이터 로딩
pop = pd.read_excel('D:/python3/data/01. population_in_Seoul.xls', encoding='utf-8')
pop.head()
```

```python
pop = pd.read_excel('D:/python3/data/01. population_in_Seoul.xls',
                    header=2, usecols='B,D,G,J,N', encoding='utf-8')
# header = 머리부분 시작위치
# usecols = 필요한 열만 골라가져옴
pop.head()  # B-자치구, D-인구(계), G-한국인(계), J-외국인(계), N-65세고령자
```

```python
# 컬럼명 변경
pop.rename(columns={pop.columns[0]:'구 별',
                   pop.columns[1]:'인구수',
                   pop.columns[2]:'한국인',
                   pop.columns[3]:'외국인',
                   pop.columns[4]:'고령자'}, inplace=True)
pop.head()
```

```python
# 기존 데이터 셋에 변수를 추가,,, 파생변수,,,?
# 파생변수 : 최근증가율
# 2013년도 이전의 데이터와 최근 3년간(14년, 15년, 16년) 증가율을 계산
cctv['최근증가율'] = (cctv['2014년']+cctv['2015년']+cctv['2016년'])/cctv['2013년도 이전']*100
cctv.head()
```

```python
# 최근 증가율을 기준으로 내림차순
cctv.sort_values(by='최근증가율', ascending=False).head()
```

```python
pop.head()
```

```python
# pop 데이터셋의 첫 행을 제거하는 작업
pop.drop([0], inplace=True)
pop.head()
```

```python
# unique() : 중복값을 제거하고 고유값을 확인하는 함수
pop['구 별'].unique()
```

```python
# 파생변수 : 외국인 비율과 고령자 비율
pop['외국인비율'] = pop['외국인'] / pop['인구수'] *100
pop['고령자비율'] = pop['고령자'] / pop['인구수'] *100
pop
```

```python
# cctv 대수와 인구수가 양의 상관관계를 가지고 있다는 것을 확인?하여 산점도 그래프에
# 상관관계를 이용하는 선을 추가하는 작업  -> 회귀선 polyfit()
fp1 = np.polyfit(df['인구수'], df['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(df['인구수'], df['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')  # ls=line style, lw=line width, g=green
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
# 오차를 계산한 후, 오차를 기준으로 정렬, 구 별 이름 추출
fp1 = np.polyfit(df['인구수'], df['소계'], 1)

f1 = np.poly1d(fp1)    # 기울기 절편
fx = np.linspace(100000, 700000, 100)

df['오차'] = np.abs(df['소계']-f1(df['인구수']))

df_sort = df.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14, 10))
plt.scatter(df['인구수'], df['소계'], c=df['오차'], s=50) # c = color
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')  # ls=line style, lw=line width, g=green

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02,  # 글자의 위치
             df_sort['소계'][n]*0.98, df_sort.index[n], fontsize=15)  
    
plt.xlabel('인구수')
plt.ylabel('인구수당 비율')
plt.grid()
plt.show()
```
