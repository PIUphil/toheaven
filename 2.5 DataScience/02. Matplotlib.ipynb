### 산점도(산포도) : plot(x, y)
- scatter()

```python
data = []
for i in range(100):
    x = np.random.normal(0, 1) # 정규분포에서 추출(0~1, 1개의 값 추출)
    y = x*0.1+0.2+np.random.normal(0, 1) # 숫자 의미 없음...?
    data.append([x, y])
```

```python
x_data = [v[0] for v in data]
#x_data
```

```python
y_data = [v[1] for v in data]
#y_data
```

```python
plt.plot(x_data, y_data, 'ro')  # r -> red / o -> 마커(동그라미)
plt.show()
```

```python
# 그림 : figure
# 한 화면(figure)에 여러개의 이미지를 담을 수 있음 - add_subplot
```

### 화면에 여러 개의 그래프를 그리기
- figure()
- add_subplot()

```python
# figure 객체
fig = plt.figure()  # figure - 도화지

# 2행x2열로 나누어 4개의 그래프를 그린다.
x1 = fig.add_subplot(2, 2, 1)  # 2x2, 화면 넘버링-> 1번영역
x2 = fig.add_subplot(2, 2, 2)  # 2번영역
x3 = fig.add_subplot(2, 2, 3)  
x4 = fig.add_subplot(2, 2, 4)

# data 생성
# 1~100 숫자 중 난수 정수를 발생, 100
data = np.random.randint(1, 100, 100) # 1~100, 100개

# 첫번째 영역
x1.hist(data)  # 히스토그램

# 두번째 영역
x2.scatter(x_data, y_data)  # 산포도(산점도)

# 세번째 영역
x3.plot(np.arange(50))  # 선그래프

# 네번째 영역
x4.plot(np.random.randn(100), 'k--')  # 선그래프  # 'b':blue, 'k':black, '--':dash(점선)

plt.show()
```

### 하나의 차트에 여러 개의 그래프를 담는 작업

```python
fig = plt.figure()
graph = fig.add_subplot(1, 1, 1)  # figure객체가 있으면 add_subplot이 반드시 있어야함

x = np.random.randn(50)
y = np.random.randn(50).cumsum()  # cumsum 누적합

# 계단형 그래프
graph.plot(x, color='r', label='step', # label='step' 범례
           drawstyle='steps-post') # drawstyle='steps' 계단형(pre,mid,post)

# 선 스타일
graph.plot(y, 'g--', label='line')  # 선그래프는 기본형이라 drawstyle 안써줘도됨

# 범례
plt.legend(loc='best')  # best - 최적(그래프에 따라 위치달라짐)
#plt.legend(loc=0) # 'best'=0, 'upper right'=1(사분면), 
                   # 오른쪽.왼쪽.아래.위.중앙=57,6,8,9,10
# 차트제목
plt.title('multi chart plotting')

# 축 레이블(축 제목)
plt.xlabel('stage')
plt.ylabel('random number')
    
plt.show()
```

```python
# 산점도 행렬 : scatter_matrix()
iris_col = iris.loc[:, 'Sepal.Length':'Petal.Width']
pd.plotting.scatter_matrix(iris_col, diagonal='kde')  # 밀도
plt.show()
```

```python
# 그래프 추세가 오른쪽 위로 올라감 - 양의 상관관계
```

```python
iris['Species'].unique()
```

```python
cols=[]
for s in iris['Species']:
    ch = 0
    if s == 'setosa':
        ch = 1
    elif s == 'versicolor':
        ch = 2
    else:
        ch = 3
    
    cols.append(ch)
    
plt.scatter(iris['Sepal.Length'], iris['Petal.Length'], c=cols)  # c = color

plt.xlabel('Sepal.Length')
plt.ylabel('Petal.Length')
plt.title('Iris Dataset')
plt.show()
```
