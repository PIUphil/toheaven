```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randn(6, 4))
df
```

```python
df.columns = ['A','B','C','D']
df.index = pd.date_range('20210501', periods=6)  # 판다스가 제공하는 날짜 형식으로 채워라
                                                 # 6개 만듦 (연속)
```

```python
df.index
```

## 결측치(NaN)

```python
df['E'] = [1.0, np.nan, 3.5, 6.1, np.nan, 7.0]
df
```

```python
df.dropna(how='all')  # 모든게 결측값인거 dropna시킴
```

```python
df.dropna(how='any') # 어느거 하나라도 결측값인거 dropna시킴
```

```python
# 결측값 대체 -> 보강법
# fillna(value=5.0)  # 값이 하나일땐 value를 안써도됨
df.fillna(5.0)
```

```python
# isnull() : 결측값입니까?  True or False  <-> notnull()
df.isnull()
```

```python
# 결측값을 가지고 있는 로우만 추출
df.loc[df.isnull()['E'], :]
```
