```python
# 바 차트 (막대그래프)
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 바 차트 : bar, barh  (bar-horizental)

y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['A', 'B', 'C']

plt.title('Bar Chart')
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(y)

# 축 제목
plt.xlabel('Word')
plt.ylabel('Fre')

plt.show()
```

```python
people = ['aaa', 'bbb', 'ccc', 'ddd']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel("x-label")

plt.show()
```

```python
# pie : 파이 차트
# http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie
# 카테고리별의 값을 상대적으로 표현,,

labels = ['aaa', 'bbb', 'ccc', 'ddd']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)   # 중심 좌표값에서 떨어져있는 비율

plt.title('Pie Chart')
plt.pie(sizes, labels=labels, colors=colors, explode=explode,
       startangle = 90, shadow=True, autopct='%1.1f%%')  # 처음에 크기를 먼저 쓰고, 나머지는 순서 중요하지않음
                        #오토퍼센트 - 소숫점 첫째자리, %기호출력 
# 파이차트를 그릴 대 원의 형태가 유지될 수 있도록 설치
plt.axis('equal')
plt.show()
```

```python
# 히스토그램 : 빈도수, hist명령어 사용,,
# 데이터를 집계할 구간 설정 - bins

x = np.random.randn(1000)

plt.title('Histogram')
plt.hist(x, bins=10)
plt.show()
```

```python
# 산점도 or 산포도 or scatter (2차원 데이터를 가지고 표현하는 그래프)
# 두 집합의 상관관계를 나타낼 때 많이 사용  /해지선,양(음)의 상관관계
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

plt.title('Scatter Plot')
plt.scatter(x, y)
plt.show()
```
