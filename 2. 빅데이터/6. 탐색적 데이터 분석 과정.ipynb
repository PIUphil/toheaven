### DataSet의 속성들
- order_id : 주문번호
- quantity : 주문수량
- item_name : 주문메뉴명

```python
# 수치적인(연속적인) 데이터의 기술통계분석
chipo.describe()
```

```python
# 주문번호는 숫자의 의미를 담지 않는다.
chipo['order_id'] = chipo['order_id'].astype(str)
```

```python
# apply() : 함수를 활용해 반복적인 작업을 효율적으로 처리
# 함수 => lambda 입력 : 출력
chipo['item_price'] = chipo['item_price'].apply(lambda x : float(x[1:]))
chipo.describe()
```

# 가장 많이 주문한 item_name의 Top10

```python
# value_counts() : 개별 요소의 건수를 추출
# value_counts는 시리즈객체만 허용되는 함수
# DataFrame['column'] 형태 : 시리즈 타입
item_count = chipo['item_name'].value_counts()[:10]
item_count
```

```python
for idx, (val, cnt) in enumerate(item_count.items()):
    print('Top', idx+1," : ", val, cnt)
```

```python
for idx, (val, cnt) in enumerate(item_count.iteritems(), 1):
    print('Top', idx," : ", val, cnt)       #인덱스가 1로 시작
```

### 주문서별 주문제품의 건수

```python
# groupby() : 특정 피처를 기준으로 그룹을 생성한 후 그룹별로 작업
# 아이템별 주문개수와 총량을 분석하는 작업
# 아이템별 주문개수
order_count = chipo.groupby('item_name')['order_id'].count()
order_count[:10]
```

```python
# 아이템별 주문수량의 총계
item_quantity = chipo.groupby('item_name')['quantity'].sum()
item_quantity[:10]
```

```python
# 아이템별 주문수량의 총량을 시각화하는 작업
# 막대그래프를 활용

item_name_list = item_quantity.index.tolist() # 리스트형으로 변형
x_pos = np.arange(len(item_name_list))
order_cnt = item_quantity.values.tolist()

plt.bar(x_pos, order_cnt, align='center') # 틱의 중앙에 막대기 놓음
plt.title('Distribution of all ordered Item')
plt.ylabel('ordered_item count')

plt.show()
```

### 각 아이템별 가격 분석 

```python
# 수량이 1개인 제품만 추출하는 작업
chipo_one_item = chipo[chipo.quantity == 1]
chipo_one_item
```

```python
# 아이템별로 그룹 후 가장 작은 금액을 추출하는 작업
price_per_item = chipo_one_item.groupby('item_name').min()
price_per_item
```

```python
price_per_item.sort_values(by='item_price')
```

```python
price_per_item.sort_values(by='item_price', ascending=False)
```

```python
price_per_item.sort_values(by='item_price')[:10]
```

```python
# 아이템별 가격 분포를 시각화하는 작업
item_name_list = price_per_item.index.tolist()
xpos = np.arange(len(item_name_list))
item_price = price_per_item['item_price'].tolist()

plt.bar(xpos, item_price, align='center')
plt.ylabel('item_price')

plt.show() # 50개 메뉴의 가격분포
```

```python
plt.hist(item_price)
plt.ylabel('count')
plt.show() # 묶음 - 가격분포
```
